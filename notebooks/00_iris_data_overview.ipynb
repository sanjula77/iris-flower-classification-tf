# Read and Display the Iris Dataset
This notebook demonstrates how to load and display the Iris dataset using pandas and scikit-learn.

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Create a DataFrame with feature data
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the target column
iris_df['target'] = iris.target

# Display the first few rows
iris_df
```

```python
# Display basic information about the dataset
iris_df.info()
```

```python
# Display summary statistics
iris_df.describe()
```

```python
# Display the distribution of target classes
iris_df['target'].value_counts()
```
